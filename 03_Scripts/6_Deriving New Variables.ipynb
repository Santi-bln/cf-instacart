# 03. Deriving new variables

```python
# define the function for price tags
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts (dropna = False)
```

```python
df['prices'].max()
```

```python
# new categories with loc() function 
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# add variable for busiest days
new_result = []

for value in ords_prods_merged["order_dow"]:
  if value == 0:
    new_result.append("Busiest_days")
  elif value == 1:
    new_result.append("Busiest_days")
  elif value == 3:
    new_result.append("Slowest_days")
  elif value ==4:
    new_result.append("Slowest_days")
  else:
    new_result.append("Regularly busy")
```

```python
updated_result = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value > 10 & value <=18:
    updated_result.append("Most_orders")
  elif value <=5:
    updated_result.append("Fewest_orders")
  else:
    updated_result.append("Average_orders")
  
```
